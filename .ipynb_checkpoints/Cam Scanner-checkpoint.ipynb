```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

im_path = 'page.jpg'
im = cv2.imread(im_path)
```

```python
print(im.shape)
img = cv2.resize(im, (1500,800))
print(img.shape)
plt.imshow(img)
plt.show()
```

```python
##image blurring(highlights the edges by detecting areas of drastic pixel changes)
orig = img.copy()
gray = cv2.cvtColor(orig, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (5,5), 0)##create a small box(kernel) on the image and scans it and normalize the pixel values
plt.imshow(blurred)
plt.show()
```

```python
##edge detecting
edged = cv2.Canny(blurred, 0, 50)
orig_edged = edged.copy()

plt.imshow(orig_edged)
plt.show()
```

```python
##contours extraction
_, contours, _ = cv2.findContours(edged, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
print(len(contours))
contours = sorted(contours, reverse=True, key=cv2.contourArea)
```

```python
##best contours
for c in contours:
    p = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, 0.02*p, True)
    
    if len(approx) == 4:
        target = approx
        break
print(target.shape)  
```

```python
##reorder target contour
## target ko 4x2 m reshape krna h aur jo points output_map m pass kre h hnew m bhi voh points ussi order m honge
def reorder(h):
    h = h.reshape(4,2)
    print(h)
    
    hnew = np.zeros((4,2), dtype=np.float32)
    
    add = h.sum(axis=1)
    hnew[0] = h[np.argmin(add)]
    hnew[2] = h[np.argmax(add)]
    
    diff = np.diff(h, axis=1)
    hnew[1] = h[np.argmin(diff)]
    hnew[3] = h[np.argmax(diff)]
    
    return hnew
```

```python
reordered = reorder(target)
print('---------')
print(reordered)
```

```python
##Project to a fixed size screen
input_represent = reordered
output_map = np.float32([[0,0], [800,0], [800,800], [0,800]])##to project image in a 800x800 window
M = cv2.getPerspectiveTransform(input_represent, output_map)
```

```python
ans = cv2.warpPerspective(orig, M, (800,800))
```

```python
plt.imshow(ans)
plt.show()
```

```python
ans2 = cv2.cvtColor(ans, cv2.COLOR_BGR2GRAY)
temp = cv2.GaussianBlur(ans2, (3,3), 0)
plt.imshow(temp)
plt.show()
```

```python
now = cv2.cvtColor(ans2, cv2.COLOR_GRAY2BGR)
plt.imshow(now)
plt.show()
```
